**Columns Selection**

Input: data = {"name": ["Alice", "Bob", "Charlie"], "age": [25, 30, 35]}, column = "age"

Output: values: [25, 30, 35] length: 3

```python
import pandas as pd

data = {"name": ["Alice", "Bob", "Charlie"], "age": [25, 30, 35]}
column = "age"
df = pd.DataFrame(data)
df.head()
```

```python
values = df[column].tolist()
values
```
